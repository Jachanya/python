```python
from IPython.display import clear_output
def printboard(board):
    for b in board:
        for item in b:
            if item not in ['x', 'o']:
                print("*   ", end = " ")
            else:
                print(item, "   ", end = " ")
        print('\n')
```

```python
def isEmpty(char):
    if char in ['x', 'o', 'X', 'O']:
        return False
    return True

def isPosEmpty(preBoard, pos):
    pos = int(pos)
    if pos == 1:
        if isEmpty(preBoard[0][0]):
            return True
        else:
            print("Position is not Empty")
            return False
    if pos == 2:
        if isEmpty(preBoard[0][1]):
            return True
        else:
            print("Position is not Empty")
            return False
    if pos == 3:
        if isEmpty(preBoard[0][2]):
            return True
        else:
            print("Position is not Empty")
            return False
    if pos == 4:
        if isEmpty(preBoard[1][0]):
            return True
        else:
            print("Position is not Empty")
            return False
    if pos == 5:
        if isEmpty(preBoard[1][1]):
            return True
        else:
            print("Position is not Empty")
            return False
    if pos == 6:
        if isEmpty(preBoard[1][2]):
            return True
        else:
            print("Position is not Empty")
            return False
    if pos == 7:
        if isEmpty(preBoard[2][0]):
            return True
        else:
            print("Position is not Empty")
            return False
    if pos == 8:
        if isEmpty(preBoard[2][1]):
            return True
        else:
            print("Position is not Empty")
            return False
    if pos == 9:
        if isEmpty(preBoard[2][2]):
            return True
        else:
            print("Position is not Empty")
            return False
```

```python
def playingBoard(preBoard,player, pos):
    '''
    PreBoard: previous state of the playing board
    player:current player
    pos: position player played
    functions return the current state of board, printed
    '''
    #print("Currently in playing board")
    pos = int(pos)
    if pos == 1:
        preBoard[0][0] = player
        return preBoard
    if pos == 2:
        preBoard[0][1] = player
    if pos == 3:
        preBoard[0][2] = player
    if pos == 4:
        preBoard[1][0] = player
    if pos == 5:
        preBoard[1][1] = player
    if pos == 6:
        preBoard[1][2] = player
    if pos == 7:
        preBoard[2][0] = player
    if pos == 8:
        preBoard[2][1] = player
    if pos == 9:
        preBoard[2][2] = player
    #print(preBoard)
    return preBoard
```

```python
def gameLogic(board):
    if board[0][0] == board[0][1] == board[0][2]:
        return True
    if board[1][0] == board[1][1] == board[1][2]:
        return True
    if board[2][0] == board[2][1] == board[2][2]:
        return True
    if board[0][0] == board[1][0] == board[2][0]:
        return True
    if board[0][1] == board[2][1] == board[2][1]:
        return True
    if board[0][2] == board[1][2] == board[2][2]:
        return True
    if board[0][0] == board[1][1] == board[2][2]:
        return True
    if board[0][2] == board[1][1] == board[2][0]:
        return True
    return False
```

```python
def inputValid(board, pos):
    if pos.isdigit():
        if isPosEmpty(board, pos):
            return True
    return False
```

```python
#Game play logic
contplay = True
i = 1
print("player1: x, player2: o")
board = [[1,2,3], [4,5,6], [7,8,9]]
printboard(board)
while contplay:
    if i % 2 == 0:
        pos = input("player2 play: ")
    else:
        pos = input("player1 play: ")
    inptv = True
    while inptv:
        inptv = not inputValid(board, pos)
        if not inputValid(board, pos):
            print("Input a valid pos")
            printboard(board)
            if i % 2 == 0:
                pos = input("player2 play: ")
            else:
                pos = input("player1 play: ")
    if i % 2 == 0:
        board = playingBoard(board, 'o', pos)
        #print(board)
    else:
        board = playingBoard(board, 'x', pos)
        #print(board)
    printboard(board)
    if gameLogic(board):
        if i % 2 == 0:
            print("Player2 wins")
            print("Player1 lose")
        else:
            print("Player1 wins")
            print("Player2 lose")
        break
        
    i += 1
    clear_output(wait =True)
```
